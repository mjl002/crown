#### As of 6/23/22, there are 383 buildings with a Costar Pull match, and 48 buildings that did not show a matching Costar Pull. So we will look at the buildings that we have not looked up on Costar yet, which is 719 buildings.

```python
df = df[df['CoStar pull'].isna()]
df.shape
```

```python
df['Full Address'] = df['Address'] +' '+ df['Zip']
df.head()
```

```python
small.head()
```

#### We change the Costar data to show only the rows with matching On-Net Buildings

```python
match = small[(small['Address'] +' '+ small['Zip'].map(str)).isin(df['Full Address'])]
match
```

#### We want to append the 'match' table to a master DF with all the finished data 

```python
master = pd.DataFrame(columns = costar.columns)
```

```python
master = master.append(match)
```

```python
master
```

#### We want to add a 'Y' to df['Costar pull']

```python
master_list = (master['Address'] +' '+ master['Zip'].map(str)).tolist()
master_list
```

```python
df['Costar pull'] = np.where(df['Full Address'].isin(master_list), 'Y', np.NaN)
```
